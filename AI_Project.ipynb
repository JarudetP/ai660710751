```python
import pandas as pd
import re

# Reload the Excel file without header to inspect full structure
print("--- ‡∏£‡∏µ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå Excel ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á ---")
df_temp = pd.read_excel(full_path, header=None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(df_temp.to_string())
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
print(f"\nüéâ ‡∏£‡∏µ‡πÇ‡∏´‡∏•‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: df_temp ‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô {len(df_temp):,} ‡πÅ‡∏ñ‡∏ß")
print(f"‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå: {df_temp.shape[1]}")

# Identify header row
N = 5 # Row 5 (index 5) contains the month headers

# Create new column names for Metric ID and Metric Name, and concatenate with month headers
new_column_names = ['Metric ID', 'Metric Name'] + list(df_temp.iloc[N, 2:])

# Create df_silver from relevant data rows and assign new column names
df_silver = df_temp.iloc[N+1:].copy()
df_silver.columns = new_column_names

# Reset the index of the DataFrame
df_silver.reset_index(drop=True, inplace=True)

# Clean column names: remove ' p' and strip whitespace
df_silver.columns = df_silver.columns.str.replace(' p', '', regex=False).str.strip()

# Drop rows where all values (excluding 'Metric ID' and 'Metric Name') are NaN
df_silver = df_silver.dropna(subset=df_silver.columns[2:], how='all').copy()

print("--- DataFrame ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏Å‡∏≥‡∏´‡∏ô‡∏î Header ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á, ‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß Metadata, ‡πÅ‡∏•‡∏∞‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå ---")
print(df_silver.head())
print(f"‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÅ‡∏ñ‡∏ß‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î: {len(df_silver):,}")
print(f"‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå: {list(df_silver.columns)}")

print("--- Melting DataFrame and Converting 'Month_Year' with Thai Month to English Conversion ---")

# 1. Use pd.melt() to transform df_silver to long format
id_vars = ['Metric ID', 'Metric Name']
value_vars = df_silver.columns.drop(id_vars)

df_melted = pd.melt(df_silver, id_vars=id_vars, value_vars=value_vars,
                    var_name='Month_Year', value_name='Number of Tourists')

# 2. Define a mapping for Thai month abbreviations to English month abbreviations
thai_month_map = {
    '‡∏°.‡∏Ñ.': 'Jan',
    '‡∏Å.‡∏û.': 'Feb',
    '‡∏°‡∏µ.‡∏Ñ.': 'Mar',
    '‡πÄ‡∏°.‡∏¢.': 'Apr',
    '‡∏û.‡∏Ñ.': 'May',
    '‡∏°‡∏¥.‡∏¢.': 'Jun',
    '‡∏Å.‡∏Ñ.': 'Jul',
    '‡∏™.‡∏Ñ.': 'Aug',
    '‡∏Å.‡∏¢.': 'Sep',
    '‡∏ï.‡∏Ñ.': 'Oct',
    '‡∏û.‡∏¢.': 'Nov',
    '‡∏ò.‡∏Ñ.': 'Dec'
}

# Function to convert Thai month-year string (BE) to English month-year string (CE)
def convert_thai_month_year_revised(month_year_str):
    if pd.isna(month_year_str):
        return None

    original_str = str(month_year_str).strip()
    parts = original_str.split(' ')
    if len(parts) == 2:
        thai_month_abbr = parts[0].strip()
        be_year_str = parts[1].strip()

        if not thai_month_abbr.endswith('.'):
            thai_month_abbr += '.'

        english_month_abbr = thai_month_map.get(thai_month_abbr)

        if english_month_abbr:
            try:
                be_year = int(be_year_str)
                ce_year = be_year - 543
                return f"{english_month_abbr} {ce_year}"
            except ValueError:
                pass

    return original_str

# Apply the conversion function to the 'Month_Year' column
df_melted['Month_Year_Converted'] = df_melted['Month_Year'].apply(convert_thai_month_year_revised)

# 3. Convert 'Month_Year_Converted' column to Datetime type
df_melted['Month_Year'] = pd.to_datetime(df_melted['Month_Year_Converted'], format='%b %Y', errors='coerce')

# Drop the helper column
df_melted.drop(columns=['Month_Year_Converted'], inplace=True)

# 4. Convert 'Number of Tourists' column to numeric type
df_melted['Number of Tourists'] = pd.to_numeric(df_melted['Number of Tourists'], errors='coerce')

print("\n--- 5 ‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡∏Ç‡∏≠‡∏á DataFrame ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£ Melt ‡πÅ‡∏•‡∏∞‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà ---")
print(df_melted.head())
print("\n--- ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏ô DataFrame ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£ Melt ‡πÅ‡∏•‡∏∞‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà ---")
print(df_melted.dtypes)
print(f"\nüéâ Melting ‡πÅ‡∏•‡∏∞‡πÅ‡∏õ‡∏•‡∏á‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: df_melted ‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô {len(df_melted):,} ‡πÅ‡∏ñ‡∏ß")

print("--- ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏î‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏Ç‡∏≠‡∏á df_melted ---")
print(df_melted.isnull().sum())

print("--- ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'Metric Name' ‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå ---\n")
df_melted['Metric Name'] = df_melted['Metric Name'].str.strip()
print("--- 5 ‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡∏Ç‡∏≠‡∏á DataFrame ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î --- ")
print(df_melted.head())
print("\n--- ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏ô DataFrame ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î ---")
print(df_melted.dtypes)
print(f"\nüéâ ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏î‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡πÅ‡∏•‡∏∞‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏ï‡∏¥‡∏°‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: df_melted ‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô {len(df_melted):,} ‡πÅ‡∏ñ‡∏ß")
```

```python
print("--- ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'Year', 'Month', ‡πÅ‡∏•‡∏∞ 'Month_Name' ---")

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà‡∏ä‡∏∑‡πà‡∏≠ 'Year'
df_melted['Year'] = df_melted['Month_Year'].dt.year

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà‡∏ä‡∏∑‡πà‡∏≠ 'Month' (‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç)
df_melted['Month'] = df_melted['Month_Year'].dt.month

# 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà‡∏ä‡∏∑‡πà‡∏≠ 'Month_Name' (‡∏ä‡∏∑‡πà‡∏≠‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡πÄ‡∏ï‡πá‡∏°)
df_melted['Month_Name'] = df_melted['Month_Year'].dt.strftime('%B')

print("\n--- 5 ‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡∏Ç‡∏≠‡∏á df_melted ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà ---")
print(df_melted.head())
print("\n--- ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏ô df_melted ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡∏°‡πà ---")
print(df_melted.dtypes)
print(f"\nüéâ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'Year', 'Month', ‡πÅ‡∏•‡∏∞ 'Month_Name' ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: df_melted ‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô {len(df_melted):,} ‡πÅ‡∏ñ‡∏ß")

print("--- ‡∏™‡∏£‡∏∏‡∏õ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ô‡∏±‡∏Å‡∏ó‡πà‡∏≠‡∏á‡πÄ‡∏ó‡∏µ‡πà‡∏¢‡∏ß‡∏£‡∏≤‡∏¢‡∏õ‡∏µ (Yearly Aggregation) ---")
annual_summary = df_melted.groupby('Year')['Number of Tourists'].sum().reset_index()
print(annual_summary.head())

print("\n--- ‡∏™‡∏£‡∏∏‡∏õ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ô‡∏±‡∏Å‡∏ó‡πà‡∏≠‡∏á‡πÄ‡∏ó‡∏µ‡πà‡∏¢‡∏ß‡∏£‡∏≤‡∏¢‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡πÅ‡∏•‡∏∞‡∏£‡∏≤‡∏¢‡∏õ‡∏µ (Monthly and Yearly Aggregation) ---")
monthly_annual_summary = df_melted.groupby(['Year', 'Month', 'Month_Name'])['Number of Tourists'].sum().reset_index()
monthly_annual_summary = monthly_annual_summary.sort_values(by=['Year', 'Month'])
print(monthly_annual_summary.head())
print(f"\nüéâ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏∏‡∏ì‡∏•‡∏±‡∏Å‡∏©‡∏ì‡∏∞‡πÅ‡∏•‡∏∞‡∏™‡∏£‡∏∏‡∏õ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: annual_summary ‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô {len(annual_summary):,} ‡πÅ‡∏ñ‡∏ß")
print(f"üéâ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏∏‡∏ì‡∏•‡∏±‡∏Å‡∏©‡∏ì‡∏∞‡πÅ‡∏•‡∏∞‡∏™‡∏£‡∏∏‡∏õ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: monthly_annual_summary ‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô {len(monthly_annual_summary):,} ‡πÅ‡∏ñ‡∏ß")

print("--- ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏≠‡∏á df_melted ---")
df_melted.info()
print("\n--- ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏≠‡∏á annual_summary ---")
annual_summary.info()
print("\n--- ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏≠‡∏á monthly_annual_summary ---")
monthly_annual_summary.info()

print("--- ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏î‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏Ç‡∏≠‡∏á df_melted ---")
print(df_melted.isnull().sum())
print("\n--- ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏î‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏Ç‡∏≠‡∏á annual_summary ---")
print(annual_summary.isnull().sum())
print("\n--- ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏î‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏Ç‡∏≠‡∏á monthly_annual_summary ---")
print(monthly_annual_summary.isnull().sum())

print("--- ‡πÅ‡∏™‡∏î‡∏á‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡πÄ‡∏ä‡∏¥‡∏á‡∏û‡∏£‡∏£‡∏ì‡∏ô‡∏≤‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÉ‡∏ô df_melted ---")
print(df_melted.describe())

print("\n--- ‡πÅ‡∏™‡∏î‡∏á‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡πÄ‡∏ä‡∏¥‡∏á‡∏û‡∏£‡∏£‡∏ì‡∏ô‡∏≤‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÉ‡∏ô annual_summary ---")
print(annual_summary.describe())
print("\n--- ‡πÅ‡∏™‡∏î‡∏á‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡πÄ‡∏ä‡∏¥‡∏á‡∏û‡∏£‡∏£‡∏ì‡∏ô‡∏≤‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÉ‡∏ô monthly_annual_summary ---")
print(monthly_annual_summary.describe())

print("\n--- 5 ‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡∏Ç‡∏≠‡∏á df_melted --- ")
print(df_melted.head())
print("\n--- 5 ‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡∏Ç‡∏≠‡∏á annual_summary --- ")
print(annual_summary.head())
print("\n--- 5 ‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡∏Ç‡∏≠‡∏á monthly_annual_summary --- ")
print(monthly_annual_summary.head())
print("\nüéâ ‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ñ‡∏∏‡∏ì‡∏†‡∏≤‡∏û‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô Gold Layer ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå.")
```
